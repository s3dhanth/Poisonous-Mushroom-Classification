## Number of Missing Values

```python
dictn = {}
for i in dataset.columns:
    dictn[i] = dataset[i].isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
missing_values = dataset.isnull().mean() * 100
missing_values = missing_values[missing_values >0]
missing_values = missing_values.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Percentage of Missing Values')
plt.title('Missing Values Distribution in df_train')
plt.show()
```

```python
pd.DataFrame(dictn.items(), columns=['columns','counts'])
```

## Mean of Missing values (gives you an approx percentage)

```python
dictn_mean = {}
for i in dataset.columns:
    dictn_mean[i] = dataset[i].isnull().mean()
pd.DataFrame(dictn_mean.items(), columns=['columns','mean']).round(1)
```

```python
Mean_analyses = pd.DataFrame(dictn_mean.items(), columns=['columns','mean']).round(1)
```

Columns with more than 60 percent of Null values (We will drop these columns)

```python
Mean_analyses[Mean_analyses['mean'] > 0.6]
```

```python
dataset.drop(columns= Mean_analyses[Mean_analyses['mean'] > 0.6]['columns'].values.tolist(), inplace= True)
```
